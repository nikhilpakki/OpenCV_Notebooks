## Introduction to OpenCV in Python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

### Read images, videos and webcam

#### Reading an image

```python
# reading an image
img = cv2.imread('cat.jpg')

# Show the image on a new window, window title as "Cat image"
cv2.imshow("Cat image", img)

# waitkey value decides how long the image window should be shown in msec, if 0 is entered then infinitely it will be open
cv2.waitKey(0)

# close all imshow windows
cv2.destroyAllWindows()
```

The **VideoCaptureObject** properties can be modified as shown below:

#### Reading a video file

```python
# opening the video file
vid = cv2.VideoCapture('HZNK3855.MOV')

while True:
    
    # reading the video frame by frame in each while loop
    status, img = vid.read()
    
    # if the image is not captured (OR) video file is complete (OR) 'q' is pressed -> break the loop
    if not status or 0xFF == ord('q'):
        break
        
    # show each frame of the video for 10 milli seconds
    cv2.imshow("Video", img)
    cv2.waitKey(10)

# close all windows
cv2.destroyAllWindows()
```

#### Capturing video from webcam

```python
WEBCAM_ID = 0 # by default the laptop webcam ID is 0, need to enter the video cam ID registered

# keycodes for 'Esc' and 'Q'
ESC_key = 27
Q_key   = 81

# opening the webcam
cam = cv2.VideoCapture(WEBCAM_ID)

while True:
    
    # reading the video frame by frame in each while loop
    status, img = cam.read()
    
    # if the image is not captured (OR) video file is complete -> break the loop
    if not status:
        break
        
    # show each frame of the video for 10 milli seconds
    cv2.imshow('Live webcam', img)
    key = cv2.waitKey(10)
    
    # 'ESC' or 'Q' is pressed -> break the loop
    if key == 27 or key == 81:
        break

# close the webcam
cam.release()
        
# close all windows
cv2.destroyAllWindows()
```

### Basic functions

#### Grayscale image

```python
img = cv2.imread('cat.jpg')

grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Gray image', grayImg)
cv2.waitKey(0)

plt.title('Gray Image')
plt.imshow(grayImg, cmap='gray')
```

#### Blurring

```python
img = cv2.imread('cat.jpg')

kernelSize = (11,11)
blurImg = cv2.GaussianBlur(img, ksize=kernelSize, sigmaX=0)

cv2.imshow('Gray image', blurImg)
cv2.waitKey(0)

plt.title('Blur Image')
plt.imshow(cv2.cvtColor(blurImg, cv2.COLOR_BGR2RGB))
```

```python
img = cv2.imread('cat.jpg')

edgesImg = cv2.Canny(img, 250, 200)

cv2.imshow('Edge detection', edgesImg)
cv2.waitKey(0)

plt.title('Edge detection')
plt.imshow(cv2.cvtColor(edgesImg, cv2.COLOR_BGR2RGB))
```

```python
kernel = np.array((3,3), np.uint8)

dilatedImg = cv2.dilate(edgesImg, kernel, iterations=1)
erodedImg = cv2.erode(edgesImg, kernel, iterations=1)

cv2.imshow('Dilated image', dilatedImg)
cv2.waitKey(0)

cv2.imshow('Eroded image', erodedImg)
cv2.waitKey(0)

plt.figure(figsize=(12,8))
plt.subplot(121)
plt.title('Dilated image')
plt.imshow(cv2.cvtColor(dilatedImg, cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.title('Eroded image')
plt.imshow(cv2.cvtColor(erodedImg, cv2.COLOR_BGR2RGB))
```

```python
img = np.zeros((512, 512, 3), np.uint8)

# ':' indicates that all the pixels of the image and R, G, B values are assigned
img[:] = 255, 0, 0

plt.figure(figsize=(10,6))
plt.subplot(131)
plt.imshow(img)
img[:] = 0, 255, 0
plt.subplot(132)
plt.imshow(img)
img[:] = 0, 0, 255
plt.subplot(133)
plt.imshow(img)
```

```python
img = np.zeros((512, 512, 3), np.uint8)

plt.figure(figsize=(10,6))
plt.subplot(131)
img = cv2.line(img, pt1=(0,0), pt2=(100,100), color=(255,0,0), thickness=3)
plt.imshow(img)
img = cv2.line(img, pt1=(0,0), pt2=(img.shape[1], img.shape[0]), color=(255,0,0), thickness=3)
plt.subplot(132)
plt.imshow(img)
img = cv2.line(img, pt1=(img.shape[0], 0), pt2=(0,img.shape[1]), color=(255,0,0), thickness=3)
plt.subplot(133)
plt.imshow(img)
```

```python
arr = np.array([[353, 202], [703, 162],  
                [942, 426], [484, 498]], np.int32)
img = cv2.imread('cards.jpg')
cv2.polylines(img, [arr], True, (0, 255, 0), 10)
plt.imshow(img)
```

```python
width, height = 250, 350
pt1 = np.float32([[353, 202], [703, 162], [484, 498], [942, 426]])
pt2 = np.float32([[0,0], [width, 0], [0, height], [width, height]])
matrix = cv2.getPerspectiveTransform(pt1, pt2)
imgOut = cv2.warpPerspective(img, matrix, (width, height))
plt.imshow(imgOut)
```

```python
arr = np.array([[111, 219], [287, 188],  
                [154, 482], [352, 440]], np.int32)
img = cv2.imread('card.jpg')
cv2.polylines(img, [arr], False, (0, 255, 0), 10)
plt.imshow(img)
```

```python
img = cv2.imread('card.jpg')
width, height = 250, 350
pt1 = np.float32([[111, 219], [287, 188], [154, 482], [352, 440]])
pt2 = np.float32([[0,0], [width, 0], [0, height], [width, height]])
matrix = cv2.getPerspectiveTransform(pt1, pt2)
imgOut = cv2.warpPerspective(img, matrix, (width, height))
plt.imshow(imgOut)
```

```python
img = cv2.imread('cat.jpg')
hStackImage = np.hstack((img, img))
vStackImage = np.vstack((img, img))
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title('Horizontal stack')
plt.imshow(cv2.cvtColor(hStackImage, cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.title('Verical stack')
plt.imshow(cv2.cvtColor(vStackImage, cv2.COLOR_BGR2RGB))
plt.show()
```
